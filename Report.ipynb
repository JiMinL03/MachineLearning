```python
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r'C:\Users\915-13\Desktop\Online Retail.xlsx')
print(df)
```

```python
df[df['Quantity'] <=0]
```

```python
df = df[df.Quantity > 0]
df
```

```python
df = df.groupby(['InvoiceNo', 'Description'])['Quantity'].sum()
df
```

```python
df = df.unstack()
df
```

```python
def encode_units(x):
    if x>=1:
        return 1
    else:
        return 0

df = df.apply(lambda x: x.map(encode_units))
print(df)
```

```python
freq=df.sum(axis=0).to_frame('Frequency')
freq_sort = freq.sort_values('Frequency', ascending=False)
freq_sort
```

```python
freq_sort['Support'] = freq_sort['Frequency'] / len(freq_sort)
freq_sort
```

```python
freq_sort['Support'].hist()
```

```python
freq_items = apriori(df, min_support=0.01, use_colnames=True, low_memory=True)
print(freq_items)
```

```python
rules = association_rules(freq_items, metric="confidence", min_threshold=0.6)
rules
```

```python
rules=rules[rules['lift']>1]
rules.sort_values(by='confidence', ascending=False)
```

```python
rules[rules.antecedents.apply(lambda x:'HERB MARKER THYME' in x)]
```

```python
rules[rules.consequents.apply(lambda x: 'HERB MARKER THYME' in x)]
```
